```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy.stats as stats
df = pd.read_csv('mydata.csv')
data={"1":df["frist"].value_counts()[1],"2":df["frist"].value_counts()[2],"3":df["frist"].value_counts()[3],"4":df["frist"].value_counts()[4],"5":df["frist"].value_counts()[5],"1 ":df["second"].value_counts()[1],"2 ":df["second"].value_counts()[2],"3 ":df["second"].value_counts()[3],"4 ":df["second"].value_counts()[4],"5 ":df["second"].value_counts()[5]," 1 ":df["third"].value_counts()[1]," 2 ":df["third"].value_counts()[2]," 3 ":df["third"].value_counts()[3]," 4 ":df["third"].value_counts()[4]," 5 ":0}
rating=list(data.keys())
values=list(data.values())

f_value,p_value=stats.f_oneway(values,rating)
print(f_value,p_value)
print("the avarage rate of productivity in the frist slot : " ,np.mean(df['frist']))
print("the madian rate of productivity in the frist slot : ",np.median(df['frist']))
print("the mode of rate of productivity in the frist slot : ",stats.mode(df['frist']))
print("the avarage rate of productivity in the second slot : ",df['second'].mean())
print("the madian rate of productivity in the second slot : ",df['second'].median())
print("the mode of rate of productivity in the second slot : ",stats.mode(df['second']))
print("the avarage rate of productivity in the third slot : ",df['third'].mean())
print("the madian rate of productivity in the third slot : ",df['third'].median())
print("the mode of rate of productivity in the third slot : ",stats.mode(df['third']))

```

```python
data={"1":df["frist"].value_counts()[1],"2":df["frist"].value_counts()[2],"3":df["frist"].value_counts()[3],"4":df["frist"].value_counts()[4],"5":df["frist"].value_counts()[5],"1 ":df["second"].value_counts()[1],"2 ":df["second"].value_counts()[2],"3 ":df["second"].value_counts()[3],"4 ":df["second"].value_counts()[4],"5 ":df["second"].value_counts()[5]," 1 ":df["third"].value_counts()[1]," 2 ":df["third"].value_counts()[2]," 3 ":df["third"].value_counts()[3]," 4 ":df["third"].value_counts()[4]," 5 ":0}
rating=list(data.keys())
values=list(data.values())
plt.xlabel("frist slot    -----    second slot    -----    thrid slot",fontsize=18)
plt.ylabel("no. of students",fontsize=16)
color=[ 'purple','purple','purple','purple','purple','green','green','green','green','green','orange','orange','orange','orange','orange']
plt.bar(rating,values,color=color)
plt.title("Student productivity through university days")
plt.show()
```

```python
cars = ['1', '2', '3','4', '5']
data = [df["frist"].value_counts()[1],df["frist"].value_counts()[2],df["frist"].value_counts()[3],df["frist"].value_counts()[4],df["frist"].value_counts()[5]]
fig = plt.figure(figsize =(10, 7))
fig, ax = plt.subplots()
plt.title("Student productivity through frist slot")
ax.pie(data, labels=cars, autopct='%1.1f%%')
plt.show()
```
